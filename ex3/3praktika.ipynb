Выгрузите страницу https://en.wikipedia.org/wiki/Python и сформируйте список ссылок, которые есть на этой странице.

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://en.wikipedia.org/wiki/Python")
soup = BeautifulSoup(response.content, "html.parser")

links_list = []

for link in soup.find_all('a'):
	href = link.get('href')
	links_list.append(href)

print(links_list)
```

1.Сделать POST запрос при помощи Python с модулем requests
2.Сделать запрос при помощи zeep

```python
import requests
# SOAP URL
url = "https://www.postman.com/cs-demo/public-soap-apis/request/34m7jic/list-of-continents-by-name"
 
# structured XML
payload = """<?xml version="1.0" encoding="utf-8"?>
<soap12:Envelope xmlns:soap12="http://www.w3.org/2003/05/soap-envelope">
  <soap12:Body>
    <ListOfContinentsByName xmlns="http://www.oorsprong.org/websamples.countryinfo">
    </ListOfContinentsByName>
  </soap12:Body>
</soap12:Envelope>"""
# headers
headers = {
    'Content-Type': 'text/xml; charset=utf-8'
}
# POST запрос
response = requests.request("POST", url, headers=headers, data=payload)
 
# ответ
print(response.text)
print(response)
```

```python
import zeep

wsdl_url = "http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso?WSDL"
client = zeep.Client(wsdl=wsdl_url)
country_code = "RU"
result = client.service.CountryIntPhoneCode(sCountryISOCode=country_code)

print(f"Country's telephone code: {country_code} - {result}")
```
